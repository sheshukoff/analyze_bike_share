# Цели анализа bike share

### 1. Разбор гипотез

#### 🔄 Ребаланс и оптимизация сети станций

- **Гипотеза 1.** Нужно посмотреть на каких станциях брали больше и оставли меньше, это поможет выявить где можно добавить количество велосипедов.
- **Гипотеза 2.** Повлияет ли уменьшение количества станций на прибыль аренды велосипедов, если убрать не ликвидные станции.
- **Гипотеза 3.** На сколько можно увеличить количество аренды велосипедов поощряя пользователей за правильную парковку.
- **Гипотеза 4.** На сколько можно увеличить прибыль если расположить парковки велосипедов рядом с торговыми точками (будто питание или сервис по велосипедам)

#### ❄️ Погодные условия

- **Гипотеза 1.** Насколько уменьшится аренда (длительность аренды) велосипедов от зимних условий, (это гололед, снег, температура), стоит ли заменить летние велосипеды на более снаряженные
- **Гипотеза 2.** Насколько уменьшится количество положительных отзывов от зимних условий (отсутствуют данные по отзывам).
- **Гипотеза 3.** По подсчетам количество дней когда идет дождь около 600, стоит ли оборудовать 30% всех велосипедов под дождь.

#### 💰 Тарифы и монетизация

- **Гипотеза 1.** Можно ли увеличить количество аренды в такие часы как с 8 до 9 и с 18 до 19 тариф на работу и домой.
- **Гипотеза 2.** Можно ли увеличить количество аренды в такие часы как с 23 до 8 позиционируя их как ночной тариф.
- **Гипотеза 3.** На сколько можно увеличить прибыль предлагая перейти на тариф member.

#### 🚀 Новые продукты и сегменты клиентов

- **Гипотеза 1.** Как можно увеличить количество аренды электровелосипедов, например для курьеров.
- **Гипотеза 2.** Можно ли увеличить количество аренды велосипедов (любых) сдавая их компании из нескольких человек (4 или более) или корпоративным клиентам.
- **Гипотеза 3.** Можно ли увеличить количество аренды велосипедов, если будет путь по разным food trip.
- **Гипотеза 4.** На сколько можно увеличить количество аренды велосипедов позиционируя их как "Bike Dating".
- **Гипотеза 5.** На сколько можно увеличить количество аренды велосипедов сдавая их как походные или как выезд на природу (от 1 до нескольких дней).

#### 👤 Поведение и сегментация пользователей

- **Гипотеза 1.** Выяснить, кто берёт велосипед на более продолжительное время — member или casual.

### Моделирование

- **Гипотеза 1.** На сколько можно увеличить прибыль если бизнес захочет открыть новую станцию
для модели нужны:
    - точки интереса
    - станции вокруг
    - суммарное количество точек интереса на станцию 

Цель — оценить потенциальную прибыль от открытия новой станции в заданной локации.
