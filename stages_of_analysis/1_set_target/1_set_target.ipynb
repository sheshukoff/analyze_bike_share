# Цели анализа bike share

### Гипотезы

- Гипотеза 1 - Нужно посмотреть на каких станциях брали больше и оставли меньше, это поможет выявить где можно добавить количество велосипедов
- Гипотеза 2 - Насколько уменьшится аренда (длительность аренды) велосипедов от зимних условий, (Это гололед, снег, температура), стоит ли заменить летние велосипеды на более снаряженные
- Гипотеза 3 - Насколько уменьшится количество положительных отзывов от зимних условий (отсутствуют данные по отзывам)
- Гипотеза 4 - По подсчетам количество дней когда идет дождь около 600, стоит ли оборудовать процентов 30 всех велосипедов под дождь
- Гипотеза 5 - Как можно увеличить количество аренды электровелосипедов, например для курьеров.
- Гипотеза 6 - Можно ли увеличить количество аренды велосипедов (любых) сдавая их компании из нескольких человек (4 или более) или корпоративным клиентам
- Гипотеза 7 - Повлияет ли уменьшение на прибыль аренды велосипедов, если убрать не ликвидные станции.
- Гипотеза 8 - Можно ли увеличить количество аренды велосипедов, если будет путь по разным food trip.
- Гипотеза 9 - Выяснить кто более продолжительно берет велосипед из (member, casual)
- Гипотеза 10 - Можно ли увеличить количество аренды в такие часы как с 8 до 9 и с 18 до 19 тариф на работу и домой
- Гипотеза 11 - Можно ли увеличить количество аренды в такие часы как с 23 до 8 позиционируя их как ночной тариф
- Гипотеза 12 - На сколько можно увеличить количество аренды велосипедов позиционируя их как Bike Dating
- Гипотеза 13 - На сколько можно увеличить количество аренды велосипедов сдавая их как походные или как выезд на природу (от 1 до нескольких дней)
- Гипотеза 14 - На сколько можно увеличить количество аренды велосипедов поощряя пользователей за правильную парковку
- Гипотеза 15 - На сколько можно увеличить прибыль предлагая перейти на тариф member
- Гипотеза 16 - На сколько можно увеличить прибыль если расположить парковки велосипедов рядом с торговыми точками (будто питание или сервис по велосипедам)

### Моделирование

На сколько можно увеличить прибыль если бизнес захочет открыть новую станцию
для модели нужны:
- данные о погоде (Можно убрать если станцию будут открывать в этом же городе)
- точки интереса
- станции вокруг
- зависимость расстояния от центра
