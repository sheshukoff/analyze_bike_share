```python
import pandas as pd
import numpy as np
```

## Сначала получим точки всех станций

- читаю файл bike_share/daily_rent_detail.csv

```python
daily_rent_detail = pd.read_csv('../../bike_share/daily_rent_detail.csv')
daily_rent_detail
```

- убирал станцию 'MTL-ECO5-03' так как она занимает 2 строчки во всем dataframe

```python
daily_rent_detail = daily_rent_detail[daily_rent_detail['start_station_id'] != 'MTL-ECO5-03']
```

- заменил в столбце start_station_id Nan на 0

```python
daily_rent_detail['start_station_id'] = daily_rent_detail['start_station_id'].replace(np.nan, 0)
```

- заменил в столбце end_station_id Nan на 0

```python
daily_rent_detail['end_station_id'] = daily_rent_detail['end_station_id'].replace(np.nan, 0)
```

- посмотрел результат

```python
daily_rent_detail
```

- оставляю только те станции, которые не равны 0

```python
daily_rent_detail = daily_rent_detail[(daily_rent_detail['start_station_id'] != 0) & (daily_rent_detail['end_station_id'] != 0)]
```

- меняю типы данных у столбцов (start_station_id, end_station_id) на float32

```python
daily_rent_detail = daily_rent_detail.astype(
    {
        'start_station_id': 'float32',
        'end_station_id': 'float32'
    }
)
```

- затем меняю типы данных (start_station_id, end_station_id) на int32 так как станции были и сплавающей точкой и целым числом

```python
daily_rent_detail = daily_rent_detail.astype(
    {
        'start_station_id': 'int32',
        'end_station_id': 'int32'
    }
)
```

- смотрю что получилось

```python
daily_rent_detail.head()
```

- написал код который получит для каждой станции медианное значение точки, где расположенна станция

```python
station_dict = {
    int(k): [float(v[0]), float(v[1])]
    for k, v in (
        daily_rent_detail
        .groupby('start_station_id')[['start_lat', 'start_lng']]
        .median()
        .apply(lambda row: [round(row['start_lat'], 6), round(row['start_lng'], 6)], axis=1)
        .to_dict()
        .items()
    )
}
station_dict
```

- читаю файл bike_share/station_list.csv

```python
station_list = pd.read_csv('../../bike_share/station_list.csv')
station_list
```

- записываю точки в dataframe

```python
station_list[['start_lat', 'start_lng']] = station_list['station_id'].map(station_dict).apply(pd.Series).fillna(0)
```

- оставляю только те станции, которые у которых есть точки

```python
station_list = station_list[(station_list['start_lat'] != 0) & (station_list['start_lng'] != 0)]
```

- смотрю что получилось

```python
station_list
```

- записываю dataframe в файл csv (data_additional/station_list.scv)

```python
station_list.to_csv('../../data_additional/station_list.scv', index=False)
```
