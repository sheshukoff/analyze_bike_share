### Описательная статистика

- данные для сезонов (в формате списка)

```python
season_order = ['winter', 'spring', 'summer', 'autumn']
```

- добавлю сезоны для столбца seasons

```python
season = {
    1: 'winter',
    2: 'winter',
    3: 'spring',
    4: 'spring',
    5: 'spring',
    6: 'summer',
    7: 'summer',
    8: 'summer',
    9: 'autumn',
    10: 'autumn',
    11: 'autumn',
    12: 'winter'
}
```

- добавляю столбец месяц

```python
weather['month'] = weather['datetime'].dt.month
```

```python
weather['seasons'] = weather['month'].map(season)
```

- делаю группировку по столбцам temp, humidity и precip и делаю агрегацию (mean, median и std)

```python
weather.groupby('seasons')[['temp', 'humidity', 'precip']].agg(['mean', 'median', 'std']).round(2).T
```

- меняю тип данных на category и задаю в какой последовательности будут идти времена года

```python
weather['seasons'] = pd.Categorical(weather['seasons'], categories=season_order, ordered=True)
```

- строю график boxplot по временам сезона (температура)

```python
fig, ax = plt.subplots(figsize=(8, 6))

weather.boxplot(column='temp', by='seasons', ax=ax, positions=range(len(season_order)))

ax.set_xticklabels(season_order)
ax.set_title('Температура по сезонам')
ax.set_xlabel('Сезон')
ax.set_ylabel('Температура, °C')
plt.suptitle('') 

plt.show()
```

- вывожу график температуры за весь период

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(
    weather['datetime'],
    weather['temp'],
    linewidth=0.7,
    color='blue'
)

ax.set_title('Температура за весь период')
ax.set_ylabel('Температура °C')

plt.show()
```

### Распеределние признаков

- хочу посмотреть распеределение трех столбцов 'conditions', 'description' и 'icon'

- для начала выведу уникальные значения conditions

```python
weather['conditions'].unique()
```

- это описание погоды (46 описаний) возможно мне не нужно будет такая конкретика

- во это точно подойдет, что посмотреть занимаемую долю погоды за весь период

```python
weather['icon'].unique()
```

- делаю группировку по столбцу icon и считаю их количество

```python
weather_icon = weather.groupby('icon')['datetime'].count().to_frame(name='count')
weather_icon
```

```python
weather_icon['share_percent'] = (weather_icon['count'] / weather_icon['count'].sum() * 100).round(2)
weather_icon
```

- выведу график barh

```python
fig, ax = plt.subplots(figsize=(10, 6))

# сортирую по возрастанию, чтобы самая большая доля была сверху
weather_icon_sorted = weather_icon.sort_values('count', ascending=True)

bars = ax.barh(
    weather_icon_sorted.index,
    weather_icon_sorted['count'],
    color='steelblue'
)

# считаю проценты и подписываю каждый столбец справа от него
total = weather_icon_sorted['count'].sum()
for bar, value in zip(bars, weather_icon_sorted['count']):
    pct = value / total * 100
    ax.text(
        bar.get_width() + total * 0.01,  # небольшой отступ от края столбца
        bar.get_y() + bar.get_height() / 2,
        f'{value} ({pct:.1f}%)',
        va='center',
        ha='left',
        fontsize=10
    )

ax.set_xlabel('Количество дней')
ax.set_title('Распределение по типам погоды')

# убираю лишние рамки для чистоты
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
